Below I show how q-q plots can be created from scratch if you're interested in the mechanics!

```python
# sort the data
roe_sort = salary.roe.sort_values()

# Divide the standard and scaled normal distirbutions into slices of equal size
# Each point in the original sample is a "quantile"
norm_quants = stats.norm.ppf(np.linspace(.001, .999, len(roe_sort)))
norm_quantsy = stats.norm(roe_sort.mean(), roe_sort.std()).ppf(np.linspace(.001, .999, len(roe_sort)))

# Plot the data
plt.scatter(norm_quants, roe_sort, color = "blue")
plt.plot(norm_quants, norm_quantsy, color = "red")
plt.title("Q-Q Plot")
plt.xlabel("Theoretical Quantiles")
plt.ylabel("Observed Values")
plt.show()
```

### Regression in Python

#### Manual

```python
# subset the x and y variables
x = salary['roe']
y = salary['salary']

# calculate the covariance and variance 
cov_xy = np.cov(x,y)[1,0] # returns a covariance matrix, we want the covariance between x and y
var_x = np.var(x, ddof = 1)

# Calculate coefficients
b1 = cov_xy/var_x
b0 = np.mean(y) - b1*np.mean(x)

print("The slope of the regression line is:", b1, "and the intercept is:", b0)
```

```python
import statsmodels.formula.api as smf
```

```python
# syntax is model = sm.OLS('y~x', data = sampl)
reg = smf.ols('salary ~ roe', data= salary)
results = reg.fit()
```

```python
# Access the estimated parameters in our results
results.params
```

```python
# the fitted values attribute contains 209 
# predicted y values for each observation
results.fittedvalues
```

```python
plt.figure(figsize = (16, 9))

# plot the observed values
plt.scatter(salary['roe'], salary['salary'])

# plot roe vs the predicted or fitted values
plt.plot(salary['roe'], results.fittedvalues, color = "orange")

# add labels
plt.title("CEO Salary vs Return on Investment")
plt.xlabel("CEO Salary (Thousands)")
plt.ylabel("ROE")

plt.plot()
```

### Residuals

```python
# calculate the fitted values from our estimated coefficients
y_hat = b0 + b1 * salary['roe']

# calculate the residuals
u_hat = salary["salary"] - y_hat

u_hat
```

```python
reg = smf.ols('salary ~ roe', data= salary)
results = reg.fit()

# residuals are also stored in results from statsmodels
results.resid
```

```python
# confirm property 1
prop1 = u_hat.mean()

# confirm property 2
prop2 = np.cov(x, u_hat)

#confirm property 3
y_barp3 = b0+b1*np.mean(x)
y_bar = np.mean(y)
```
